Примените магическую функцию %matplotlib inline для отображения графиков в Jupyter Notebook и настройки конфигурации ноутбука со значением 'svg' для более четкого отображения графиков. Создайте список под названием x с числами 1, 2, 3, 4, 5, 6, 7 и список y с числами 3.5, 3.8, 4.2, 4.5, 5, 5.5, 7. С помощью функции plot постройте график, соединяющий линиями точки с горизонтальными координатами из списка x и вертикальными из списка y.
```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

##### Список x:

```python
x = np.arange(1, 8)
print(x)
```

##### Список y:

```python
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
print(y)
```

##### Строим график y(x), в котором точки соединены линиями:

```python
plt.plot(x,y)
plt.show()
```

##### Строим график y(x), в котором точки не соединены линиями:

```python
plt.scatter(x,y)
plt.show()
```
